```python
# 1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.
def sum_even_numbers(numbers):
  sum_of_evens = 0
  for number in numbers:
    if number % 2 == 0:
      sum_of_evens += number
  return sum_of_evens
# Example usage
numbers = [1, 2, 3, 4, 5, 6]
even_sum = sum_even_numbers(numbers)
print(f"Sum of even numbers: {even_sum}")
```

```python
# 4. Write a Python function that checks if a given number is prime or not from 1 to 200.
def is_prime(number):
  if number <= 1:
    return False
  for i in range(2, int(number**0.5) + 1):
    if number % i == 0:
      return False
  return True

# Example usage
for number in range(1, 201):
  if is_prime(number):
    print(f"{number} is a prime number")
```

```python
#5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
class FibonacciIterator:
    def __init__(self, n):
        self.n = n
        self.a = 0
        self.b = 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count < self.n:
            result = self.a
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return result
        else:
            raise StopIteration

# Example usage
fib_iterator = FibonacciIterator(10)
for number in fib_iterator:
    print(number)
```

```python
# 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
def powers_of_two(exponent):
    for i in range(exponent + 1):
        yield 2**i
# Example usage
for power in powers_of_two(5):
    print(power)
```

```python
# 7. Implement a generator function that reads a file line by line and yields each line as a string.
def file_reader(filename):
    try:
        with open(filename,'w') as file:
            file.write("My name is upkar sharma\n")
            file.write("I am a software er\n")
            file.write("I am a java developer and now learning python also\n")
            file.close()


        with open(filename, 'r') as file:
            for line in file:
                yield line.strip()
    except FileNotFoundError:
        print(f"Error: File '{filename}' not found.")
#Example usage: Assuming you have a file named 'my_file.txt'
for line in file_reader('my_file.txt'): #replace my_file.txt with your file
    print(line)
```

```python
#10. Create a Python program that uses `filter()` to remove all the vowels from a given string.
def remove_vowels(string):
    vowels = "aeiouAEIOU"
    return "".join(filter(lambda char: char not in vowels, string))

input_string = "Hello, World!"
string_without_vowels = remove_vowels(input_string)
print(string_without_vowels) #Output: Hll, Wrld!
```

```python
#11) Imagine an accounting routine used in a book shop. It works on a list with sublists, which look like this:
orders = [ [1, ("5464", 4, 9.99), ("8274",18,12.99), ("9744", 9, 44.95)],
          [2, ("5464", 9, 9.99), ("9744", 9, 44.95)],
          [3, ("5464", 9, 9.99), ("88112", 11, 24.99)],
          [4, ("8732", 7, 11.99), ("7733",11,18.99), ("88112", 5, 39.95)] ]

from functools import reduce

order_tuples = list(map(lambda x: (x[0], list(map(lambda y: y[1]*y[2], x[1:]))), orders))

final_result = []
for order in order_tuples:
  order_num = order[0]
  product_sum = reduce(lambda x, y: x + y, order[1])

  if product_sum < 100:
      final_result.append((order_num, product_sum + 10))
  else:
      final_result.append((order_num, product_sum))

final_result
```
