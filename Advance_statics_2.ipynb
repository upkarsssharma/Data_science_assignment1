```python
1. What is hypothesis testing in statistics?
Answer:
Hypothesis testing is a statistical method used to make decisions or inferences about population parameters based on sample data. It involves formulating two competing hypotheses and using sample evidence to determine which is more likely.
```

```python
2. What is the null hypothesis, and how does it differ from the alternative hypothesis?
Answer:

Null Hypothesis (H₀): Assumes no effect or no difference; it's the default or status quo hypothesis.

Alternative Hypothesis (H₁ or Ha): Suggests there is an effect or a difference.

Example:

H₀: μ = 50 (The mean is 50)

H₁: μ ≠ 50 (The mean is not 50)
```

```python
3. What is the significance level in hypothesis testing, and why is it important?
Answer:
The significance level (α) is the probability of rejecting the null hypothesis when it is actually true (Type I error). Common values are 0.05 or 0.01. It sets the threshold for deciding whether an observed effect is statistically significant.
```

```python
4. What does a P-value represent in hypothesis testing?
Answer:
A P-value is the probability of obtaining test results at least as extreme as the observed results, assuming the null hypothesis is true.
```

```python
5. How do you interpret the P-value in hypothesis testing?
Answer:

If P-value ≤ α: Reject the null hypothesis (evidence supports the alternative).

If P-value > α: Do not reject the null hypothesis (insufficient evidence).
```

```python
6. What are Type I and Type II errors in hypothesis testing?
Answer:

Type I Error (False Positive): Rejecting H₀ when it is true (probability = α).

Type II Error (False Negative): Not rejecting H₀ when it is false (probability = β).
```

```python
7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
Answer:

One-tailed test: Tests for effect in one direction (e.g., H₁: μ > 50).

Two-tailed test: Tests for effect in both directions (e.g., H₁: μ ≠ 50).
```

```python
8. What is the Z-test, and when is it used in hypothesis testing?
Answer:
A Z-test is used when the population standard deviation is known and the sample size is large (n ≥ 30). It tests hypotheses about population means or proportions.
```

```python
9. How do you calculate the Z-score, and what does it represent in hypothesis testing?
Answer:
Z-score formula:
z = (x̄ - μ) / (σ / √n)
x̄: sample mean

μ: population mean

σ: population standard deviation

n: sample size

The Z-score tells how many standard deviations the sample mean is from the population mean.
```

```python
10. What is the T-distribution, and when should it be used instead of the normal distribution?
Answer:
The T-distribution is used instead of the normal distribution when the sample size is small (n < 30) and the population standard deviation is unknown. It accounts for more variability.
```

```python
11. What is the difference between a Z-test and a T-test?
Answer:

Criteria	Z-test	T-test
Population σ known?	Yes	No
Sample size	Large (n ≥ 30)	Small (n < 30)
Distribution	Normal	T-distribution
```

```python
12. What is the T-test, and how is it used in hypothesis testing?
Answer:
A T-test is used to compare sample means when the population standard deviation is unknown. Types include:

One-sample t-test

Independent two-sample t-test

Paired sample t-test
```

```python
13. What is the relationship between Z-test and T-test in hypothesis testing?
Answer:
Both tests assess the difference between means. A T-test is used when σ is unknown and/or the sample size is small. A Z-test is preferred when σ is known and the sample size is large.
```

```python
14. What is a confidence interval, and how is it used to interpret statistical results?
Answer:
A confidence interval (CI) gives a range of values within which the true population parameter is likely to lie with a certain confidence level (e.g., 95%).
```

```python
15. What is the margin of error, and how does it affect the confidence interval?
Answer:
The margin of error (ME) determines the width of the confidence interval. Larger ME → wider interval → less precision. It depends on the standard deviation, sample size, and confidence level.
```

```python
16. How is Bayes' Theorem used in statistics, and what is its significance?
Answer:
Bayes' Theorem updates the probability of a hypothesis based on new evidence:

P(A|B) = [P(B|A) * P(A)] / P(B)
It is significant in decision-making under uncertainty, like spam filtering or medical diagnosis.
```

```python
17. What is the Chi-square distribution, and when is it used?
Answer:
The Chi-square (χ²) distribution is used for categorical data to assess how expected counts compare to observed counts. It is right-skewed and used in goodness-of-fit and independence tests.
```

```python
18. What is the Chi-square goodness of fit test, and how is it applied?
Answer:
This test evaluates whether observed frequencies match expected frequencies.
Formula:
χ² = Σ((O - E)² / E)
O = observed frequency

E = expected frequency
```

```python
19. What is the F-distribution, and when is it used in hypothesis testing?
Answer:
The F-distribution compares two variances. It’s used in the ANOVA test and F-tests to determine if the variability between groups is more than within groups.
```

```python
20. What is an ANOVA test, and what are its assumptions?
Answer:
ANOVA (Analysis of Variance) tests whether three or more group means are significantly different.
Assumptions:

Independence

Normality

Homogeneity of variance
```

```python
21. What are the different types of ANOVA tests?
Answer:

One-Way ANOVA: One independent variable

Two-Way ANOVA: Two independent variables

Repeated Measures ANOVA: Same subjects across conditions
```

```python
22. What is the F-test, and how does it relate to hypothesis testing?
Answer:
An F-test compares two variances and is the basis of ANOVA. It helps determine whether observed variances are statistically significantly different.
```

```python
import numpy as np
from scipy import stats

# Sample data
sample_data = [78, 85, 92, 88, 75, 80, 84, 90, 86, 89]

# Population parameters
population_mean = 82
population_std = 5

# Z-test calculation
sample_mean = np.mean(sample_data)
n = len(sample_data)
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(n))
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

print(f"Z-Score = {z_score:.4f}")
print(f"P-Value = {p_value:.4f}")
if p_value < 0.05:
    print("Result: Reject the null hypothesis.")
else:
    print("Result: Fail to reject the null hypothesis.")
```

```python
3. Q: Implement a one-sample Z-test using Python to compare the sample mean with the population mean.
A:
```

```python
def one_sample_z_test(data, pop_mean, pop_std):
    sample_mean = np.mean(data)
    n = len(data)
    z = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))
    p = 2 * (1 - stats.norm.cdf(abs(z)))
    return z, p

# Example usage:
sample = [50, 52, 48, 53, 51, 49, 54]
z, p = one_sample_z_test(sample, 50, 3)
print(f"Z = {z:.4f}, P = {p:.4f}")
```

```python
5. Q: Create a Python function that calculates and visualizes Type I and Type II errors during hypothesis testing.
A:
```

```python
def visualize_errors(mu0=100, mu1=105, sigma=10, n=30, alpha=0.05):
    se = sigma / np.sqrt(n)
    z_alpha = stats.norm.ppf(1 - alpha)
    x = np.linspace(90, 115, 1000)

    null_dist = stats.norm(mu0, se)
    alt_dist = stats.norm(mu1, se)

    plt.plot(x, null_dist.pdf(x), label="Null Hypothesis (H₀)", color='blue')
    plt.plot(x, alt_dist.pdf(x), label="Alternative Hypothesis (H₁)", color='green')

    critical_value = mu0 + z_alpha * se
    plt.axvline(critical_value, color='red', linestyle='--', label="Critical Value")

    # Type I Error (α): reject H₀ when H₀ is true
    plt.fill_between(x, null_dist.pdf(x), where=(x > critical_value), color='red', alpha=0.3, label='Type I Error')

    # Type II Error (β): fail to reject H₀ when H₁ is true
    plt.fill_between(x, alt_dist.pdf(x), where=(x <= critical_value), color='orange', alpha=0.3, label='Type II Error')

    plt.title('Type I and Type II Errors')
    plt.legend()
    plt.grid(True)
    plt.show()

visualize_errors()
```

```python
9. Q: Write a Python function to calculate the confidence interval for a sample mean and explain its significance.
A:
```

```python
def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data)
    margin = std_err * stats.t.ppf((1 + confidence) / 2, df=n-1)
    return (mean - margin, mean + margin)

sample = np.random.normal(100, 15, 30)
ci = confidence_interval(sample)
print(f"95% Confidence Interval: {ci}")
# Interpretation: This interval contains the true population mean with 95% confidence.
```

```python
10. Q: Write a Python program to calculate the margin of error for a given confidence level using sample data.
A:
```

```python
def margin_of_error(data, confidence=0.95):
    n = len(data)
    stderr = stats.sem(data)
    margin = stderr * stats.t.ppf((1 + confidence) / 2, df=n-1)
    return margin

sample_data = np.random.normal(100, 15, 30)
moe = margin_of_error(sample_data)
print(f"Margin of Error: {moe:.4f}")
```

```python
def bayes_theorem(prior_A, prob_B_given_A, prob_B):
    return (prob_B_given_A * prior_A) / prob_B

posterior = bayes_theorem(0.01, 0.95, 0.01*0.95 + 0.99*0.05)
print(f"Posterior Probability: {posterior:.4f}")
```

```python
🧠 Interpretation:
If a disease has a 1% prior chance and the test is 95% accurate, the posterior probability of having the disease after a positive test is only ~16.1%.
```

```python
from scipy.stats import shapiro, levene

def check_anova_assumptions(groups):
    normality = [shapiro(group)[1] > 0.05 for group in groups]
    equal_var = levene(*groups)[1] > 0.05
    return all(normality), equal_var
```

```python
✅ All assumptions are satisfied for using ANOVA.
```

```python
25. Q: Write a Python script to perform a Z-test for comparing proportions between two datasets or groups.
```

```python
# Sample counts
success_a, size_a = 45, 100
success_b, size_b = 30, 100

p1, p2 = success_a / size_a, success_b / size_b
p_combined = (success_a + success_b) / (size_a + size_b)
z = (p1 - p2) / np.sqrt(p_combined * (1 - p_combined) * (1/size_a + 1/size_b))
p_val = 2 * (1 - stats.norm.cdf(abs(z)))
print(f"Z = {z:.4f}, P-value = {p_val:.4f}")
```
